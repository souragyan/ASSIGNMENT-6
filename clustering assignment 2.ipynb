```python
# import hierarchical clustering libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
air=pd.read_csv('airlines.csv')
```

```python
air
```

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func(air.iloc[:,1:])
```

```python
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='average'))
```

```python
hc = AgglomerativeClustering(n_clusters=6, affinity = 'euclidean', linkage = 'single')
```

```python
y_hc = hc.fit_predict(df_norm)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
air['h_clusterid'] = hc.labels_
```

```python
#for kmeans
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
air=pd.read_csv('airlines.csv')
```

```python
air
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_air_df = scaler.fit_transform(air.iloc[:,1:])
```

```python
scaled_air_df
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_air_df)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.cluster import KMeans
clusters_new = KMeans(6, random_state=42)
clusters_new.fit(scaled_air_df)
```

```python
air['clusterid_new'] = clusters_new.labels_
```

```python
clusters_new.cluster_centers_
```
